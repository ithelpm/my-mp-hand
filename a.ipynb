```python
# -*- coding: utf-8 -*-

import matplotlib.pyplot as plt

# 定義資料
x = [0, 20, 40, 60, 80, 100]
y = [10.96, 11.29, 12.48, 13.3, 14.51, 16.18]

# 繪製圖形
plt.plot(x, y, label='輸出電壓與 VR1 設定關係曲線')

# 設定標題和軸標籤
plt.title('降壓式直流電源調整器輸出電壓與 VR1 設定關係')
plt.xlabel('VR1 (%)')
plt.ylabel('輸出電壓 (V)')

# 顯示圖形
plt.show()
```
